# Import Packages

```python
import numpy as np
import pandas as pd
```

# Dataset

```python
X = pd.DataFrame({
    "age": [25, 42, 37, 51, 29, 46, 33, 60],
    "income": [55_000, 92_000, np.nan, 120_000, 64_000, 110_000, 72_000, 130_000],
    "city": ["Houston", "Dallas", "Houston", "Austin", "Dallas", "Austin", None, "Houston"],
    "tenure_months": [3, 48, 18, 72, 8, 60, 24, 96],
})
y = pd.Series([0, 1, 0, 1, 0, 1, 0, 1], name="converted")
```

# Practice

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scalar = StandardScaler()
X_scaled = scalar.fit_transform(X[["age", "tenure_months"]])
print(scalar.mean_)

model = LogisticRegression()
model.fit(X_scaled, y)
labels = model.predict(X_scaled)
proba = model.predict_proba(X_scaled)
print(model.classes_, model.coef_)
```
